# Data Collection

```python
# Collect the Research Data. 

import requests
from bs4 import BeautifulSoup

# URL of the web page
url = "https://suninme.org/true-story"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the web page
soup = BeautifulSoup(response.content, "html.parser")

# Find all the links (anchor tags) in the web page
links = soup.find_all("a")

# Extract and print the href attribute of each link
for link in links:
    href = link.get("href")
    if href:
        print(href)

```

```python
# Extract and print the href attribute of each link with the text true-story in the URL
link_set = set() 
for link in links:
    href = link.get("href")
    if href and "true-story" in href and "twitter" not in href and "facebook" not in href:
        print(href) 
        link_set.add( href ) 
```

```python
import os
import time
import requests
from bs4 import BeautifulSoup

# Function to fetch content from a link, extract text, and save it to a file
def fetch_and_save_content(link, index):
    # Make a request to the link
    response = requests.get(link)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the web page
        soup = BeautifulSoup(response.content, "html.parser")
        # Find the page-content div by id
        page_content_div = soup.find(id="page-content")
        
        # If page-content div is found, find the first div with the specified class
        if page_content_div:
            row_div = page_content_div.find("div", class_="row")
            
            if row_div:
                target_div = row_div.find("div")
                
                if target_div:
                    # Extract the text content or other elements from the target div
                    text = target_div.get_text(separator="\n")
                    # Save the text to a file
                    file_path = os.path.join("files", f"{index}.txt")
                    with open(file_path, "w", encoding="utf-8") as file:
                        file.write(text)
                        print(f"Data written to: {file_path}")

                else:
                    print("No div with specified class found inside row div.")
            else:
                print("No row div found inside page-content div.")
        else:
            print("No page-content div found.")

    else:
        print(f"Failed to fetch content from link {index}. Status code: {response.status_code}")


```

```python
# Number of iterations
list_of_links = list( link_set ) 
max_iterations = 20 

# Iterate over the links
for i, link in enumerate(list_of_links, start=1):
    # Stop at 20 iterations
    if i > max_iterations:
        break
    
    # Fetch and save content from the link
    print(f"Working on {link}") 
    fetch_and_save_content(link, i)
    
    # Take a sleep of 10 seconds before processing the next link
    time.sleep(10)
```
